apply, map, applymap, replace

```python
# apply on series
df_people['email'].apply(len)
```

```python
def clean_email(email):
    return email.lower()
# simple example
```

```python
df_people['email'].apply(clean_email)
```

```python
# to apply on series
df_people['email'] = df_people['email'].apply(clean_email)
```

```python
# with lambda function, works on series
df_people['email'] = df_people['email'].apply(lambda x: x.lower())
```

```python
# to apply on dataframe
df_people['email'] = df_people['email'].apply(clean_email)
```
